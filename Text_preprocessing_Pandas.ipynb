```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/bank.csv", sep=";")
```

```python
df.head()
```

```python
def replace_material(val):
  if val == "single":
    return 0
  else:
    return 1
```

```python
df['marital'] = df['marital'].apply(replace_material)
```

```python
df.head()
```

```python
df["housing"] = df["housing"].map({"no": 0, "yes": 1}.get)
```

```python
df.head()
```

```python
df.tail()
```

```python
df["loan"] = df["loan"].replace({"no": 0, "yes": 1})
```

```python
df.head()
```

```python
df["job"].unique()
```

```python
df["job"].replace({
    'unknown': np.nan,
    'management': 0,
    'technician': 1,
    'entrepreneur': 2,
    'blue-collar': 3,
    'retired': 4,
    'admin.': 5,
    'services': 6,
    'self-employed': 7,
    'unemployed': 8,
    'housemaid': 9,
    'student': 10
}, inplace=True)
```

```python
df.tail()
```

```python
df["education"].replace({
    'unknown': np.nan,
    'tertiary': 0, 
    'secondary': 1, 
    'primary': 2
}, inplace=True)
```

```python
df["default"].replace({"no": 0,"yes": 1})
```
