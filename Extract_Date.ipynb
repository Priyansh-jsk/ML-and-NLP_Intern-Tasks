```python
import re
import datetime
```

```python
import pandas as pd
data = pd.read_excel('Easyocr_ImagetoText2.xlsx')
```

```python
data.head()
```

```python
def checking_valid_dates(dt):
  result = re.findall(
        r'\d{4}-\d{2}-\d{2}', dt)
  return result
```

```python
data['valid_date'] = data['Text'].apply(lambda dt: checking_valid_dates(dt))
```

```python
def checking_valid_dates2(my):
  result2 = re.findall(
        r'\d{2}-\d{4}', my)
  return result2
```

```python
data['valid_date2'] = data['Text'].apply(lambda my: checking_valid_dates2(my))
```
